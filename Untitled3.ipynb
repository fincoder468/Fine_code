```python
# create a library class
# define methods
# to display all books
# borrow a book- (who owns the book if not present in library)
# donate a book add book
# return book
# input-library name,list of books

# dictionary books-nameofperson
# create a func and run an infinite while loop asking users for their input

dic={}
class lib:
  def __init__(self,name,liss):
    self.name=name
    self.liss=liss
  
  def display(self):
    return self.liss
  
  def borrow(self,person,bookname):
    try:
      x=self.liss.index(bookname)
      dic.update({bookname:person})
      self.liss.remove(bookname)
    except ValueError:
      print(dic[bookname])
  
  def add(self,bookname):
    self.liss.append(bookname)

  def retur(self,bookname):
    self.liss.append(bookname)
    del dic[bookname]

print("give a name to your library")
y=input()
print("give the list of books")
lis=[]
while True:
  t=input()
  if t=="done":
    break
  lis.append(t)

cus=lib(y,lis)
print(cus.display())
print(f"welcome to online library {y}")
while True:
  print("please choose a option")
  print("press 1 to borrow a book")
  print("press 2 to return a book")
  print("press 3 to add or donate a book")
  ans=int(input())
  if ans=="done":
    break
  print("enter your name")
  nn=input()
  print("enter book name")
  bn=input()
  if ans==1:
    cus.borrow(nn,bn)

  elif ans==3:
    cus.add(bn)

  else:
    cus.retur(bn)
  
```
